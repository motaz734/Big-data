```python
import pandas as pd
```

```python
df = pd.read_csv("tweets.csv")
```

```python
df.head()
```

```python
import numpy as np
df_hashtags = df['hashtags'].str.split(',', expand=True)
df_hashtags = df_hashtags.replace('[]', np.nan)
df_hashtags = df_hashtags.replace(r'[\[\]]', '', regex=True)
```

```python
df_hashtags.dropna(how='all', inplace=True)
df_hashtags
```

```python


df_comb = pd.DataFrame()
temp = []
val_count_tmp = pd.Series(dtype=float)

# sum all df_hashtags in dataframe with iteration
for col in df_hashtags.columns:
    temp.append(df_hashtags[col].fillna(0).value_counts())

for val_count in temp:
    val_count_tmp = val_count_tmp.add(val_count,fill_value=0)

y = val_count_tmp.dropna().drop(0)
y.sort_values(ascending=False, inplace=True)

y = y[y>50]

y.head(10).append(y.tail(10))
```

```python

y.plot.bar(figsize=(200,100), fontsize=150)
```
