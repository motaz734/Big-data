```python
import pandas as pd
```

```python
df = pd.read_csv("tweets.csv")
```

```python
df_language = df['language']
df_language.dropna(how='all', inplace=True)
```

```python
df_language
```

```python
count_series=df_language.value_counts()
count_series = count_series[count_series > 40]
count_series.sort_values(ascending=False, inplace=True)

count_series
```

```python
index = count_series.index
count = count_series.values


df_language = pd.DataFrame({'language': index, 'count': count})
df_language 
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
count_series.plot.bar(figsize=(100,100),fontsize=70)
```
